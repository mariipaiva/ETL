```python
# Importando o pandas
import pandas as pd

#Criando a função para extração dos dados necessários, ajuste das colunas e consolidação dos dados em arquivo csv
#Trocar apenas o caminho do arquivo a cada mês

def read_files(path_json, path_csv):
    empresaA = pd.read_json(path_json)
    empresaB = pd.read_csv(path_csv)
    return empresaA, empresaB

#Identificando os cabeçalhos
def unify_headers(empresaA, empresaB):
    # Adicionar a coluna 'Data da Venda' com valor 'Indisponível' se não existir
    if 'Data da Venda' not in empresaA.columns:
        empresaA['Data da Venda'] = 'Indisponível'
    
    # Verificar se os cabeçalhos são iguais
    if list(empresaA.columns) != list(empresaB.columns):
        # Renomear as colunas do DataFrame empresaB para que correspondam às do DataFrame empresaA
        empresaB.columns = empresaA.columns
    return empresaA, empresaB

def consolidar_dataframes(empresaA, empresaB): 
    # Concatenar os DataFrames verticalmente (um embaixo do outro) 
    dados_consolidados = pd.concat([empresaA, empresaB], ignore_index=True) 
    return dados_consolidados

# Caminhos dos arquivos
path_json = '../data_raw/dados_empresaA.json'
path_csv = '../data_raw/dados_empresaB.csv'

# Ler os arquivos
empresaA, empresaB = read_files(path_json, path_csv)

# Unificar cabeçalhos
empresaA, empresaB = unify_headers(empresaA, empresaB)


# Concatenar DataFrames 
dados_consolidados = consolidar_dataframes(empresaA, empresaB) 

#Verificando através da qtde de linhas se tudo foi consolidado corretamente
print('Qtde linhas esperada: ', len(empresaA) + len(empresaB))
print('Qtde linhas Consolidado:', len(dados_consolidados))

# Salvar o resultado em um novo arquivo CSV 
dados_consolidados.to_csv('../data_processed/dados_consolidados.csv', index=False, encoding='utf-8')

# Abrindo DataFrame consolidado
dados_consolidados
```
